```python
import pandas as pd
from sklearn.model_selection import train_test_split


sk_comments_df = pd.read_csv('comments.csv', low_memory=False)

#first split the dataset into training and test
train_comments, test_comments = train_test_split(sk_comments_df, test_size=0.2)

print(type(sk_comments_df['0'][0]))
sk_comments_df.head()
```

```python
#pre-process steps to be performed on the training and test data separately
df_to_pre_process = train_comments.copy()


import string
df_columns = df_to_pre_process.columns

for column in df_columns:
    c = df_to_pre_process[column]
    
    #convert to lower-case and remove punctuation
    df_to_pre_process[column] = c.str.lower()
    translator = str.maketrans('', '', string.punctuation)
    df_to_pre_process[column] = c.str.translate(translator)


df_to_pre_process.tail()

#remove punctuation

# translator = str.maketrans('', '', string.punctuation)
# string_name=string_name.translate(translator)

#remove stop words
```
